```python
close_price = nalco_data[['Close']]
```

```python
max(close_price.values),min(close_price.values)
```

```python
from sklearn.preprocessing import  MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
scaled_data =  scaler.fit_transform(close_price)
```

```python
scaled_data
```

```python
len(scaled_data)
```

```python
x_data = []
y_data = []
for i in range(100,len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])
import numpy as np
x_data , y_data = np.array(x_data),np.array(y_data)
```

```python
x_data[0],y_data[0]
```

```python
int(len(x_data)*0.7)
```

```python
4935-100-int(len(x_data)*0.7)
```

```python
splitting_len = int(len(x_data)*0.7)
x_train = x_data[:splitting_len]
y_train = y_data[:splitting_len]
x_test = x_data[splitting_len:]
y_test = y_data[splitting_len:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python

from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Input

# Define the model
model = Sequential()

# Use Input layer to specify input shape
model.add(Input(shape=(x_train.shape[1], 1)))  # First layer with input shape
model.add(LSTM(128, return_sequences=True))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer = 'adam',loss = 'mean_squared_error')
```

```python
model.fit(x_train,y_train,batch_size=1,epochs= 2)
```

```python
model.summary()
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
inv_predictions = scaler.inverse_transform(predictions)
inv_predictions
```

```python
inv_y_test = scaler.inverse_transform(y_test)
inv_y_test
```

```python
rmse = np.sqrt(np.mean((inv_predictions-inv_y_test)**2))
rmse
```

```python
plotting_data = pd.DataFrame (
    {
        'original_test_data':inv_y_test.reshape(-1),
        'predictions':inv_predictions.reshape(-1)
    },
    index = nalco_data.index[splitting_len+100:]
)
plotting_data
```

```python
plot_graph((15,6),plotting_data,'test data')
```

```python
 plot_graph((15,6),pd.concat([close_price[:splitting_len+100],plotting_data],axis=0),'test data')
```
